### Mean Encoding

Mean encoding represents a probability of your target variable, conditional on each value of the feature. 

```python
import pandas as pd
df = pd.read_csv('titanic.csv', usecols=['cabin','survived'])
df.head()
```

```python
# replace '?' value with word 'Missing'
df['cabin'] = df['cabin'].replace('?', 'Missing')
df.head(20)
```

```python
# Get first letter for each cabin category
df['cabin'] = df['cabin'].astype(str).str[0]
df.head()
```

```python
# check all unique values for cabin feature
df['cabin'].unique()
```

```python
# apply mean for each category for cabin
mean_ordinal = df.groupby(['cabin'])['survived'].mean().to_dict()
mean_ordinal
```

```python
# map each cabin category with it's mean
df['mean_ordinal_encode'] = df['cabin'].map(mean_ordinal)
df.head(10)
```
