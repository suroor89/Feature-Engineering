```
# import all necessory packages
import pandas as pd
import numpy as np
```

```
# Load data from csv file
data = pd.read_csv("mercedesbenz.csv", usecols=["X1","X2","X3","X4","X5","X6"])
```

```
data.head()
```

```
# check how many unique labels have in each column
for col in data.columns:
  print(col, ': ', len(data[col].unique()), 'labels')
```

```
# examine the number of columns after one hot coding with pandas in build method
pd.get_dummies(data, drop_first=True).shape
```

##### Here, we are getting 117 new variables from 6 initial categorical variables. Instaed, We can select top 10 most frequent categories from each variable.

```
# Let's find top 10 categories from 'X2' variable
data['X1'].value_counts().sort_values(ascending=False).head(10)
```

```
# Let's make a list of top 10 categories of X2 variable
X1_top_10 = [col for col in data['X1'].value_counts().sort_values(ascending=False).head(10).index]
X1_top_10
```

```
# Let's make 10 binary values for each catogory of X2 variable
for label in X1_top_10:
  data[label] = np.where(data['X1']==label, 1, 0)

data[X1_top_10].head(20)
```

```
# get top 10 for each variable
X2_top_10 = [col for col in data['X2'].value_counts().sort_values(ascending=False).head(10).index]
X3_top_10 = [col for col in data['X3'].value_counts().sort_values(ascending=False).head(10).index]
X4_top_10 = [col for col in data['X4'].value_counts().sort_values(ascending=False).head(10).index]
X5_top_10 = [col for col in data['X5'].value_counts().sort_values(ascending=False).head(10).index]
X6_top_10 = [col for col in data['X6'].value_counts().sort_values(ascending=False).head(10).index]
```

```
# get whle set of dummy variables, for all categorical variables
    # function to create dummy variables for most frequest labels
    # we can vary the number of most frequent labels that we encode
def one_hot_top_x(df, variable, top_x_labels):
  for label in top_x_labels:
    data[variable+'_'+label] = np.where(data[variable]==label, 1, 0)
```

```
# Encode 'X1' variable into 10 most frequent categories
one_hot_top_x(data, 'X1', X1_top_10)
data.head()
```

```
# Drop extra columns
cols = ['aa',	's', 'b',	'l',	'v',	'r', 'i',	'a',	'c',	'o']
data.drop(cols, axis=1, inplace=True)
```

```
data.head()
```

```
# transform categorical variables into one hot encoding for remaining variables
one_hot_top_x(data, 'X2', X2_top_10)
one_hot_top_x(data, 'X3', X3_top_10)
one_hot_top_x(data, 'X4', X4_top_10)
one_hot_top_x(data, 'X5', X5_top_10)
one_hot_top_x(data, 'X6', X6_top_10)
```

```
data.head()
```

```
# drop intial categorical variables
cols=['X1', 'X2', 'X3', 'X4', 'X5', 'X6']
data.drop(cols, axis=1, inplace=True)
data.head()
```
